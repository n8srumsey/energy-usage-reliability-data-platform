```python
from pyspark.sql import functions as F

CATALOG = "energy_data_platform_project"
GOLD_SCHEMA = f"{CATALOG}.gold"
```

```python
months = spark.range(1, 13).withColumnRenamed("id", "month")

season = (
    F.when(F.col("month").isin(12, 1, 2), "winter")
     .when(F.col("month").isin(3, 4, 5), "spring")
     .when(F.col("month").isin(6, 7, 8), "summer")
     .otherwise("fall")
)
```

```python
dim_month_season = (
    months
    .withColumn("season", season)
    .withColumn(
        "month_name",
        F.date_format(F.to_date(F.col("month").cast("string"), "M"), "MMM")
    )
)
```

```python
(
    dim_month_season
    .write
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable(f"{GOLD_SCHEMA}.dim_month_season")
)
```
