```python
from pyspark.sql import functions as F

CATALOG = "energy_data_platform_project"
SILVER_SCHEMA = f"{CATALOG}.silver"
GOLD_SCHEMA = f"{CATALOG}.gold"
```

```python
weather = spark.table(f"{SILVER_SCHEMA}.noaa_isd_ksfo_clean")

dim_station = (
    weather
    .select(
        "station_id",
        "usaf",
        "wban",
        "latitude",
        "longitude",
        "elevation_m"
    )
    .dropDuplicates()
)
```

```python
(
    dim_station
    .write
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable(f"{GOLD_SCHEMA}.dim_station")
)
display(spark.table(f"{GOLD_SCHEMA}.dim_station").limit(10))
```
