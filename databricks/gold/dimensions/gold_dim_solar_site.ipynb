```python
from pyspark.sql import functions as F

CATALOG = "energy_usage_data_platform"
SILVER_SCHEMA = f"{CATALOG}.silver"
GOLD_SCHEMA = f"{CATALOG}.gold"
```

```python
solar = spark.table(f"{SILVER_SCHEMA}.ca_solar_irradiance_clean")

dim_solar_site = (
    solar
    .select("solar_site_id")
    .dropDuplicates()
)
```

```python
(
    dim_solar_site
    .write
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable(f"{GOLD_SCHEMA}.dim_solar_site")
)
display(spark.table(f"{GOLD_SCHEMA}.dim_solar_site").limit(10))
```
